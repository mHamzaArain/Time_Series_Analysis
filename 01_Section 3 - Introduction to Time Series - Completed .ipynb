### The QQ Plot

```python
import scipy.stats
import pylab
```

```python
scipy.stats.probplot(df_comp.spx, plot = pylab)
plt.title("QQ Plot", size = 24)
pylab.show()

# QQ (Qyuatile-Quantile) plot: Tells that data follows normal dist.
# X > How far standard deviation is away from mean
# Y > Values arranges in ascending manner
# Red line > If data is tracing in that line

## As we can see at 
## Std. Dev= -2
## values = 500
## shows that it is not following normal distribution.
```

```python
scipy.stats.probplot(df_comp.ftse, plot = pylab)
plt.title("QQ Plot", size = 24)
pylab.show()
```
